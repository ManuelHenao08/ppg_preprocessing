# PPG Photopletismography Processing for Heart Rate & Heart Rate Variability

Signal Description: The pulse oximetry signal (SPO2), allows the calculation of the oxygen saturation level in the blood, and it is acquired over the index finger of the patient. Under normal conditions, the variations in SPO2 have correlated with heart rhythm and its maximum value is in phase with the R wave in electrocardiographic signal (ECG).[3][5]
